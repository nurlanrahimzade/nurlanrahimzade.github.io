```python
# Calculate total sales by region
# Analyzing the sales by region:
# Calculate the total sales for each region.
# Visualize the sales by region using a bar chart.



sales_by_region = data.groupby('Region')['Sales'].sum()

# Visualize sales by region
plt.figure(figsize=(10, 6))
sales_by_region.plot(kind='bar', color='blue')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales by Region')
plt.xticks(rotation=0)
plt.show()
```

```python
# Analyzing the sales by category and sub-category:
# Calculate the total sales for each category and sub-category.
# Visualize the sales by category and sub-category using a bar chart.

# Calculate total sales by category and sub-category
sales_by_category = data.groupby(['Category', 'Sub-Category'])['Sales'].sum().unstack()

# Visualize sales by category and sub-category
plt.figure(figsize=(12, 6))
sales_by_category.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category and Sub-Category')
plt.xticks(rotation=45)
plt.legend(title='Sub-Category')
plt.show()
```

# Sales and profitability trends over time. 

```python
# Convert the Order Date column to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Extract the year and month from the Order Date
data['YearMonth'] = data['Order Date'].dt.to_period('M').astype(str)

# Calculate the monthly sales and profit
monthly_sales = data.groupby('YearMonth')['Sales'].sum()
monthly_profit = data.groupby('YearMonth')['Profit'].sum()

# Plotting the sales trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.xlabel('Year-Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()

# Plotting the profit trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_profit.index, monthly_profit.values, marker='o', color='green')
plt.xlabel('Year-Month')
plt.ylabel('Profit')
plt.title('Monthly Profit Trend')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top products in terms of sales

# Group data by product name and calculate total sales
product_sales = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Plotting the top products by sales
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.values, y=product_sales.index, orient='h')
plt.xlabel('Sales')
plt.ylabel('Product Name')
plt.title('Top Products by Sales')
plt.show()
```
